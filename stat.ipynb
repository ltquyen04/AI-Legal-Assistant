```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Đọc file
df = pd.read_csv("qa.csv")

# Làm sạch và tách tag (các tag cách nhau bằng dấu phẩy)
all_tags = []
for tags in df['tag'].dropna():
    split_tags = [t.strip() for t in tags.split(",") if t.strip()]
    all_tags.extend(split_tags)

# Thống kê tần suất
tag_counts = Counter(all_tags)
tag_df = pd.DataFrame(tag_counts.items(), columns=["Tag", "Count"]).sort_values("Count", ascending=False)

# Vẽ biểu đồ cột (top 10)
plt.figure(figsize=(10, 6))
sns.barplot(data=tag_df.head(10), x="Count", y="Tag", palette="viridis")
plt.title("🔍 Top 10 từ khoá (tag) phổ biến trong Q&A")
plt.xlabel("Số lượng câu hỏi")
plt.ylabel("Từ khoá")
plt.tight_layout()
plt.show()

# Vẽ biểu đồ tròn (pie chart top 5)
top_tags = tag_df.head(5)
plt.figure(figsize=(6, 6))
plt.pie(top_tags["Count"], labels=top_tags["Tag"], autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("📊 Tỷ lệ các từ khoá phổ biến nhất")
plt.tight_layout()
plt.show()
```

```python
# Đường dẫn đến file CSV
CSV_PATH = "qa.csv"

# Từ khoá bạn muốn thống kê
KEYWORDS = ["tạm vắng", "tạm trú", "thường trú", "cư trú", "lưu trú"]

# Đọc file CSV
df = pd.read_csv(CSV_PATH)

# Tách tag và lọc chỉ những tag nào chứa từ khoá
filtered_tags = []

for tags in df["tag"].dropna():
    for tag in tags.split(","):
        tag = tag.strip().lower()
        if any(keyword in tag for keyword in KEYWORDS):
            filtered_tags.append(tag)

# In các tag đã lọc mỗi tag một dòng và không có trùng lặp
unique_filtered_tags = set(filtered_tags)
for tag in unique_filtered_tags:
    print(tag)

```

```python
# Đếm số tag chứa từ khoá tạm vắng
special_tags = []
for tags in df["tag"].dropna():
    for tag in tags.split(","):
        tag = tag.strip().lower()
        if "cư trú" in tag:
            special_tags.append(tag)

# In các tag chứa từ khoá tạm vắng mỗi tag một dòng và không có trùng lặp
unique_special_tags = set(special_tags)
for tag in unique_special_tags:
    print(tag)
```

```python
# Thống kê số lượng từng tag
tag_counts = Counter(filtered_tags)
tag_df = pd.DataFrame(tag_counts.items(), columns=["Tag", "Count"]).sort_values("Count", ascending=True)

# Vẽ biểu đồ bar chart nằm ngang
plt.figure(figsize=(10, 6))
plt.barh(tag_df["Tag"], tag_df["Count"], color="mediumseagreen")
plt.xlabel("Số lượng câu hỏi")
plt.title("📊 Thống kê các tag chứa từ khoá cư trú")
plt.grid(axis='x', linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
