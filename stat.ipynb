```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# c file
df = pd.read_csv("qa.csv")

# Lm sch v tch tag (cc tag cch nhau bng du phy)
all_tags = []
for tags in df['tag'].dropna():
    split_tags = [t.strip() for t in tags.split(",") if t.strip()]
    all_tags.extend(split_tags)

# Thng k tn sut
tag_counts = Counter(all_tags)
tag_df = pd.DataFrame(tag_counts.items(), columns=["Tag", "Count"]).sort_values("Count", ascending=False)

# V biu  ct (top 10)
plt.figure(figsize=(10, 6))
sns.barplot(data=tag_df.head(10), x="Count", y="Tag", palette="viridis")
plt.title(" Top 10 t kho (tag) ph bin trong Q&A")
plt.xlabel("S lng cu hi")
plt.ylabel("T kho")
plt.tight_layout()
plt.show()

# V biu  trn (pie chart top 5)
top_tags = tag_df.head(5)
plt.figure(figsize=(6, 6))
plt.pie(top_tags["Count"], labels=top_tags["Tag"], autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title(" T l cc t kho ph bin nht")
plt.tight_layout()
plt.show()
```

```python
# ng dn n file CSV
CSV_PATH = "qa.csv"

# T kho bn mun thng k
KEYWORDS = ["tm vng", "tm tr", "thng tr", "c tr", "lu tr"]

# c file CSV
df = pd.read_csv(CSV_PATH)

# Tch tag v lc ch nhng tag no cha t kho
filtered_tags = []

for tags in df["tag"].dropna():
    for tag in tags.split(","):
        tag = tag.strip().lower()
        if any(keyword in tag for keyword in KEYWORDS):
            filtered_tags.append(tag)

# In cc tag  lc mi tag mt dng v khng c trng lp
unique_filtered_tags = set(filtered_tags)
for tag in unique_filtered_tags:
    print(tag)

```

```python
# m s tag cha t kho tm vng
special_tags = []
for tags in df["tag"].dropna():
    for tag in tags.split(","):
        tag = tag.strip().lower()
        if "c tr" in tag:
            special_tags.append(tag)

# In cc tag cha t kho tm vng mi tag mt dng v khng c trng lp
unique_special_tags = set(special_tags)
for tag in unique_special_tags:
    print(tag)
```

```python
# Thng k s lng tng tag
tag_counts = Counter(filtered_tags)
tag_df = pd.DataFrame(tag_counts.items(), columns=["Tag", "Count"]).sort_values("Count", ascending=True)

# V biu  bar chart nm ngang
plt.figure(figsize=(10, 6))
plt.barh(tag_df["Tag"], tag_df["Count"], color="mediumseagreen")
plt.xlabel("S lng cu hi")
plt.title(" Thng k cc tag cha t kho c tr")
plt.grid(axis='x', linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
